```python
# List all tables in bronze schema (including cleaned versions)
tables = spark.sql("SHOW TABLES IN 01_bronze_catalog.bronze_schema").collect()

print("\n" + "=" * 60)
print("ALL TABLES IN BRONZE SCHEMA")
print("=" * 60)

raw_tables = []
clean_tables = []

for table in tables:
    table_name = table.tableName
    if 'clean' in table_name:
        clean_tables.append(table_name)
    else:
        raw_tables.append(table_name)

print("\nRAW TABLES (with Fivetran metadata):")
for t in sorted(raw_tables):
    print(f"  • {t}")

print("\nCLEANED TABLES (metadata dropped, underscores removed):")
for t in sorted(clean_tables):
    print(f"  ✓ {t}")

print(f"\nTotal: {len(raw_tables)} raw tables, {len(clean_tables)} cleaned tables")
```
