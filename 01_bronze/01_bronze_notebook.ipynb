```python
def clean_bronze_table(source_table, target_table):
    """
    Drop Fivetran metadata columns and remove underscores from column names
    """
    
    df = spark.read.table(source_table)
    
    fivetran_cols = ['_file', '_line', '_modified', '_fivetran_synced']
    for col in fivetran_cols:
        if col in df.columns:
            df = df.drop(col)
    
   
    for old_col in df.columns:
        new_col = old_col.replace('_', '')
        df = df.withColumnRenamed(old_col, new_col)
    
   
    df.write.mode("overwrite").saveAsTable(target_table)
    print(f"✓ Processed: {target_table}")
    return df


tables_to_clean = [
    ("01_bronze_catalog.bronze_schema.crm_cust_base_raw", "01_bronze_catalog.bronze_stage_schema.crm_cust_base_clean"),
    ("01_bronze_catalog.bronze_schema.inv_levels_raw", "01_bronze_catalog.bronze_stage_schema.inv_levels_clean"),
    ("01_bronze_catalog.bronze_schema.mst_product_master", "01_bronze_catalog.bronze_stage_schema.mst_product_master_clean"),
    ("01_bronze_catalog.bronze_schema.raw_sales_dtl", "01_bronze_catalog.bronze_stage_schema.raw_sales_dtl_clean"),
    ("01_bronze_catalog.bronze_schema.rtn_trans_logs", "01_bronze_catalog.bronze_stage_schema.rtn_trans_logs_clean"),
    ("01_bronze_catalog.bronze_schema.stores_geo_list_final", "01_bronze_catalog.bronze_stage_schema.stores_geo_list_final_clean")
]


print("Starting bronze table cleanup...\n")
for source, target in tables_to_clean:
    clean_bronze_table(source, target)

print("\n All bronze tables cleaned successfully!")
```
